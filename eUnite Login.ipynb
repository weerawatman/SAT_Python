```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# กำหนด Username และ Password ของคุณ
username = "weerawat.m@somboon.co.th"
password = "WwMm@123456789"

# ระบุ Path ไปยัง WebDriver ของคุณ
webdriver_path = 'path/to/your/chromedriver'  # หรือ geckodriver

# เริ่มต้น WebDriver (เลือกเบราว์เซอร์ที่คุณต้องการ)
driver = webdriver.Chrome(executable_path=webdriver_path)  # สำหรับ Chrome
# driver = webdriver.Firefox(executable_path=webdriver_path) # สำหรับ Firefox

try:
    # 1) เข้าสู่เว็บไซต์
    driver.get("https://signin.eunite.com/eunite/login?account=sbg")
    print("เข้าสู่หน้า Login สำเร็จ")

    # รอให้ Element ของ Username ปรากฏ
    username_field = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "username"))
    )

    # 2) กรอก Username
    username_field.send_keys(username)
    print(f"กรอก Username: {username}")

    # รอให้ Element ของ Password ปรากฏ
    password_field = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "password"))
    )

    # กรอก Password
    password_field.send_keys(password)
    print(f"กรอก Password: {password}")

    # คลิกปุ่ม Login (อาจต้องตรวจสอบ ID หรือ Class ของปุ่ม)
    login_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, "kc-login"))  # ตรวจสอบ ID ของปุ่ม Login
    )
    login_button.click()
    print("คลิกปุ่ม Login")

    # รอสักครู่เพื่อให้หน้าเว็บโหลดหลังจาก Login (ปรับเวลาตามความเหมาะสม)
    time.sleep(5)
    print("Login อัตโนมัติสำเร็จ")

except Exception as e:
    print(f"เกิดข้อผิดพลาด: {e}")

finally:
    # ปิดเบราว์เซอร์เมื่อเสร็จสิ้น
    driver.quit()
    print("ปิดเบราว์เซอร์")
```
